```python
def normalize_comment(text, normalization_dict):
    words = text.split()
    normalized_words = [normalization_dict.get(word, word) for word in words]
    return ' '.join(normalized_words)

df['comment'] = df['comment'].apply(lambda x: normalize_comment(x, normalization_dict))
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factory = StopWordRemoverFactory()
stopwords_sastrawi = factory.get_stop_words()

stopwords_sastrawi
```

```python
def remove_stopwords(text):
    words = text.split()
    filtered = [word for word in words if word not in stopwords_sastrawi]
    return ' '.join(filtered)

df['comment_stopword'] = df['comment'].apply(remove_stopwords)
```

```python
stopwords_nltk = set(stopwords.words('indonesian'))

def remove_nltk_stopwords(text):
    return ' '.join([word for word in text.split() if word.lower() not in stopwords_nltk])

df['comment_stopword'] = df['comment_stopword'].apply(remove_nltk_stopwords)
```

```python
custom_stopwords = set(['aja', 'dong', 'nih', 'deh', 'kok', 'si', 'lah', 'nya',  'lu', 'loe', 'gue', 'banget',
                        'ma', 'emang', 'pak', 'kau', 'mah', 'yaa', 'beliau', 'mu', 'eh', 'ini', 'itu', 'lagi', 'tek',
                        'beliau', 'ayo', 'ni', 'sih', 'yah', 'gimana' 'tuh', 'kali', 'jadi', 'enggak', 'gak', 'gimana', 'nih', 'aku', 'ya' 'gitu', 'saja', 'tuh', 'doang', 'kek'])

def remove_custom_stopwords(text):
    return ' '.join([word for word in text.split() if word.lower() not in custom_stopwords])

df['comment_stopword'] = df['comment_stopword'].apply(remove_custom_stopwords)
```

```python
from tqdm import tqdm

sentiments = []

for comment in tqdm(df['comment'], desc ="Klasifikasi Sentimen"):
  try:
    result = pipe(comment)[0]
    sentiments.append(result['label'])
  except:
    sentiments.append("error")

df['sentiment'] = sentiments
```
